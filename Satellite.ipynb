```python
import math
def localgravity(theta):
  theta = math.radians(theta)
  ans = 978032.68 * (1 + 0.0053024 * (math.sin(theta)**2) - 0.0000058 * (math.sin(2 * theta)**2))
  return ans

print(localgravity(30))
```
